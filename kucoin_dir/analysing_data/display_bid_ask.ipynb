```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load the JSON data
with open('/root/trading_systems/kucoin_dir/kucoin_release_data_initial/2024-12-02_10-00_QUILL/QUILL_level1_data.json') as f:
    data = json.load(f)

# Extract the relevant data
timestamps = []
ask_prices = []
bid_prices = []

for entry in data['data']:
    timestamps.append(entry['time_received'])
    ask_prices.append(float(entry['asks'][0]))
    bid_prices.append(float(entry['bids'][0]))

# Create a DataFrame
df = pd.DataFrame({
    'timestamp': pd.to_datetime(timestamps),
    'ask_price': ask_prices,
    'bid_price': bid_prices
})
df = df.head(10)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['ask_price'], label='Ask Price', color='red')
plt.plot(df['timestamp'], df['bid_price'], label='Bid Price', color='blue')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Ask and Bid Prices Over Time')
plt.legend()
plt.show()
```
