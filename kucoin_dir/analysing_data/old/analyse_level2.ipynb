```python
import datetime
def convert(timestamp):
    # Determine the length of the timestamp to identify its unit
    length = len(str(int(timestamp)))

    if length > 18:
        # Nanoseconds
        time_in_seconds = int(timestamp) / 1_000_000_000_000
    elif length > 15:
        # Microseconds
        time_in_seconds = int(timestamp) / 1_000_000_000
    elif length > 12:
        # Milliseconds
        time_in_seconds = int(timestamp) / 1_000
    else:
        # Seconds
        time_in_seconds = int(timestamp)

    # Convert to a datetime object
    human_readable_time = datetime.datetime.fromtimestamp(time_in_seconds)
    readable_time = human_readable_time.strftime('%H:%M:%S.%f')

    return readable_time

# Convert the 'time' column to a string
df['time'] = df['time'].astype(str)
# Apply the conversion function if ':' is not in any of the values
if not df['time'].str.contains(':').any():
    df['time'] = df['time'].apply(convert)

#convert changes column to dict 
# Convert each list of tuples to a dictionary using map
df['changes'] = df['changes'].map(dict)
print(type(df.changes))

print(df.info())
df  
```

```python
changes_column = df['changes'].iloc[0]

changes_column

def check_ask(row):
    if row['changes']['asks']:
        return row['changes']['asks']

def check_bid(row):
    if row['changes']['bids']:
        return row['changes']['bids']


df['asks_change'] = df.apply(check_ask, axis=1)
df['bids_change'] = df.apply(check_bid, axis=1)
df.rename(columns={'time':'timestamp'}, inplace=True)
df_ba = df.copy()
df
df_ba
```
